```
import nltk
import csv
import numpy as np
```

```
negative = []
with open("words_negative.csv", "r") as wn:
  reader = csv.reader(wn)
  for row in reader:
    negative.append(row)
```

```
positive = []
with open("words_positive.csv", "r") as wp:
  reader = csv.reader(wp)
  for row in reader:
    positive.append(row)
```

```
positive[:10]
```

```
negative[:10]
```

```
def sentiment(text):
  sents = nltk.sent_tokenize(text)
  for sent in sents:
    ncount = 0
    pcount = 0
    words = nltk.word_tokenize(sent)
    for word in words:
      for item in positive:
        if(word == item[0]):
          pcount += 1
      for item in negative:
        if(word == item[0]):
          ncount += 1

    if(pcount > 0 and ncount == 0):
      print("+ : ", sent)

    elif(ncount % 2 > 0):
      print("- : ", sent)

    elif(ncount % 2 == 0 and ncount > 0):
      print("+ : ", sent)

    else:
      print("? : ",sent)
```

```
def sentiment_value(text):
  sents = nltk.sent_tokenize(text)
  temp = []
  for sent in sents:
    ncount = 0
    pcount = 0
    words = nltk.word_tokenize(sent)
    for word in words:
      for item in positive:
        if(word == item[0]):
          pcount += 1
      for item in negative:
        if(word == item[0]):
          ncount += 1

    if(pcount > 0 and ncount == 0):
      temp.append(1)

    elif(ncount % 2 > 0):
      temp.append(-1)

    elif(ncount % 2 == 0 and ncount > 0):
      temp.append(1)

    else:
      temp.append(0)

  return temp
```

```
comments = []
with open("reviews.csv", "r") as file:
  reader = csv.reader(file)
  for row in reader:
    comments.append(row)
```

```
comments
```

```
for review in comments:
  print("\n")
  print(np.average(sentiment_value(str(review))))
  print(review)
```
