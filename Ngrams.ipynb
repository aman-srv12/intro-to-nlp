```
import nltk
```

```
nltk.download("all")
```

```
text = "My name is Aman. I study in Stony Brook University"
```

```
tokens = nltk.word_tokenize(text)
```

```
bigrams = nltk.bigrams(tokens)
```

```
for item in bigrams:
  print (item)
```

```
trigrams = nltk.trigrams(tokens)
```

```
for item in trigrams:
  print (item)
```

```
from nltk.util import ngrams
```

```
text2 = "My name is Aman. I work at Nutanix Inc. I study MS in CS at Stony Brook University"
```

```
tokens = nltk.word_tokenize(text2)
```

```
bigram = ngrams(tokens, 2)
```

```
for item in bigram:
  print(item)
```

```
def n_grams(text, n):
  tokens = nltk.word_tokenize(text)
  grams = ngrams(tokens, n)
  return grams
```

```
out = n_grams(text2, 4)
```

```
for item in out:
  print(item)
```
