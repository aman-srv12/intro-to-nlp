```
import nltk
nltk.download("all")
```

```
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```
alice = [word.lower() for word in alice if word.isalpha()]
```

```
fd = nltk.FreqDist(alice)
```

```
fd_100 = fd.most_common(100)
```

```
words = [word[0] for word in fd_100]
```

```
words
```

```
out = set(words) - set(nltk.corpus.stopwords.words("english"))
```
