```
norm = [word.lower() for word in md if word.isalpha()]
```

```
md_pos = nltk.pos_tag(norm, tagset = "universal")
```

```
md_pos[:5]
```

```
md_nouns = [word[0] for word in md_pos if word[1] == "NOUN"]
```

```
md_nouns[:10]
```

```
md_nouns_fd = nltk.FreqDist(md_nouns)
```

```
md_10 = md_nouns_fd.most_common(10)
```

```
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```
tokens = [word.lower() for word in alice if word.isalpha()]
```

```
pos = nltk.pos_tag(tokens, tagset="universal")
```

```
for token in pos:
  if token[0] == 'over' or token[0] == 'spoke' or token[0] == 'answer':
    print(token) 
```

```
alice_fd = nltk.ConditionalFreqDist(pos)
```

```
alice_fd
```

```
alice_fd["over"]
```

```
alice_fd["spoke"]
```

```
alice_fd["answer"]
```

```
stories = nltk.corpus.gutenberg.words("bryant-stories.txt")
```

```
tags = nltk.pos_tag(stories, tagset = "universal")
```

```
for ((word1, tag1), (word2, tag2), (word3, tag3))in nltk.trigrams(tags):
  if tag1 == "NOUN" and word2 == "or" and tag3 == "NOUN":
    print(word1 + " " + word2 + " " + word3)
```
